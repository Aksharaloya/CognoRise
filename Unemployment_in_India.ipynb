```python
plt.figure(figsize=(15,10))
sns.countplot(y="Region",data=df1)
plt.show()
```

```python
avg_unemployment_rate = df1.groupby('Region')[" Estimated Unemployment Rate (%)"].mean() # removed extra space from column name
state_with_highest_unemployment = avg_unemployment_rate.idxmax()
high_unemployment_rate = avg_unemployment_rate.max()
state_with_lowest_unemployment = avg_unemployment_rate.idxmin()
low_unemploy_rate = avg_unemployment_rate.min()
print(f"state with high employment : {state_with_highest_unemployment}")
print(f"high employment rate : {high_unemployment_rate}")
print(f"state with low employment : {state_with_lowest_unemployment}")
print(f"low employment rate : {low_unemploy_rate}")
```

```python
regions = avg_unemployment_rate.index
avg_rates = avg_unemployment_rate.values

plt.figure(figsize=(10, 6))
plt.bar(regions, avg_rates, color='skyblue')
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate')
plt.title('Average Unemployment Rate by Region')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Region', y=' Estimated Unemployment Rate (%)', data=df1,palette="Set1") # added a missing space in front of the column name
plt.title("Unemployment Rate Distribution by Region")
plt.xlabel("State")
plt.ylabel("Estimated Unemployment Rate")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.histplot(data=df1, x=' Estimated Unemployment Rate (%)', kde=True,color="navy") # removed extra spaces from column name and added a missing space at the beginning
plt.show()
```

```python
plt.figure(figsize=(14, 7))
# Create the box plot with separate colors for each region
sns.boxplot(x='Region', y=' Estimated Unemployment Rate (%)', data=df1)
# Rotate x-axis labels for better readability
plt.xticks(rotation=90)
# Add title and axis labels
plt.title('Box Plot of Estimated Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10,5))
df1.groupby('Region')[' Estimated Unemployment Rate (%)'].mean().plot(kind='pie', autopct='%1.1f%%')
plt.title('Area wise Unemployment Rate')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
corr = df1[[' Estimated Unemployment Rate (%)', ' Estimated Employed', ' Estimated Labour Participation Rate (%)']].corr()
sns.heatmap(corr, annot=True, linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
avg_unemployement_over_time = df1.groupby(' Date')[' Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False).reset_index()
print("Average Unemployment Rate over time\n", avg_unemployement_over_time)
```

```python
avg_unemployement_over_region = df1.groupby('Region')[' Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False).reset_index()
print("\nAverage Unemployment Rate over Region\n", avg_unemployement_over_region)
```

```python
avg_people_employed_over_region = df1.groupby('Region')[' Estimated Employed'].mean().sort_values(ascending=False).reset_index()
print("\nAverage People employed over Region\n", avg_people_employed_over_region)
```

```python
vg_labour_part_rate_over_region = df1.groupby('Region')[' Estimated Labour Participation Rate (%)'].mean().sort_values(ascending=False).reset_index()
print("\nAverage Labour Participation Rate over Region\n", vg_labour_part_rate_over_region) # Changed avg_labour_part_rate_over_region to vg_labour_part_rate_over_region
```

```python
avg_unemployement_area_wise = df1.groupby('Region')[' Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False).reset_index()
print("\nAverage Unemployment Rate over Area\n", avg_unemployement_area_wise)
```

```python
highest_unemployement_rate_month = df1.groupby(' Date')[' Estimated Unemployment Rate (%)'].max().sort_values(ascending=False).reset_index()
print("\nHighest Unemployment Rate Month wise\n", highest_unemployement_rate_month)
```
